```python
import pandas as pd
import numpy as np
import pyarrow as pyarrow
import requests
import os
```

```python
from sqlalchemy import create_engine, text

engine = create_engine(
    "postgresql://postgres:000@localhost:5432/tlc_dw"
)

conn = engine.connect()
print(conn)
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS datetime_dim (
    datetime_id INTEGER PRIMARY KEY,
    full_datetime TIMESTAMP NOT NULL,
    year INTEGER,
    month INTEGER,
    day INTEGER,
    hour INTEGER,
    minute INTEGER,
    weekday INTEGER,
    weekday_name TEXT,
    is_weekend BOOLEAN
);
"""))
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS location_dim (
    location_id INTEGER PRIMARY KEY,
    borough TEXT,
    zone TEXT,
    service_zone TEXT
);
"""))
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS payment_dim (
    payment_type_id INTEGER PRIMARY KEY,
    payment_description TEXT
);
"""))
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS vendor_dim (
    vendor_id INTEGER PRIMARY KEY,
    vendor_name TEXT
);
"""))
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS ratecode_dim (
    ratecode_id INTEGER PRIMARY KEY,
    ratecode_description TEXT
);
"""))
```

```python
conn.execute(text("""
CREATE TABLE IF NOT EXISTS trips_fact (
    trip_id BIGSERIAL PRIMARY KEY,

    pickup_datetime_id INTEGER REFERENCES datetime_dim(datetime_id),
    dropoff_datetime_id INTEGER REFERENCES datetime_dim(datetime_id),

    pickup_location_id INTEGER REFERENCES location_dim(location_id),
    dropoff_location_id INTEGER REFERENCES location_dim(location_id),

    payment_type_id INTEGER REFERENCES payment_dim(payment_type_id),
    vendor_id INTEGER REFERENCES vendor_dim(vendor_id),
    ratecode_id INTEGER REFERENCES ratecode_dim(ratecode_id),

    passenger_count FLOAT,
    trip_distance FLOAT,
    fare_amount FLOAT,
    extra FLOAT,
    mta_tax FLOAT,
    tip_amount FLOAT,
    tolls_amount FLOAT,
    improvement_surcharge FLOAT,
    total_amount FLOAT,
    congestion_surcharge FLOAT,
    airport_fee FLOAT,
    cbd_congestion_fee FLOAT,
    store_and_fwd_flag CHAR(1)
);
"""))
conn.commit()
```

```python
pd.read_sql("""
SELECT table_name
FROM information_schema.tables
WHERE table_schema='public';
""", engine)
```

```python
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data"
os.makedirs("data/raw", exist_ok=True)

for month in range(1, 3):
    mm = str(month).zfill(2)
    filename = f"yellow_tripdata_2025-{mm}.parquet"
    url = f"{base_url}/{filename}"
    path = f"data/raw/{filename}"

    r = requests.get(url, stream=True, timeout=60)

    if r.status_code == 200:
        with open(path, "wb") as f:
            for chunk in r.iter_content(8192):
                f.write(chunk)
        print(f"Downloaded {filename}")
    else:
        print(f"Missing {filename}")
```

```python
df = pd.read_parquet("data/raw/yellow_tripdata_2025-01.parquet")
df.head()
```

```python
all_datetimes = pd.concat([
    df["tpep_pickup_datetime"],
    df["tpep_dropoff_datetime"]
])
all_datetimes = all_datetimes.drop_duplicates()
datetime_dim = pd.DataFrame({
    "datetime_id": range(1, len(all_datetimes) + 1),
    "full_datetime": all_datetimes
})

datetime_dim = datetime_dim.reset_index(drop=True)
datetime_dim
```

```python
df.isnull().sum()
```

```python
datetime_dim["year"] = datetime_dim["full_datetime"].dt.year
datetime_dim["month"] = datetime_dim["full_datetime"].dt.month
datetime_dim["day"] = datetime_dim["full_datetime"].dt.day
datetime_dim["hour"] = datetime_dim["full_datetime"].dt.hour
datetime_dim["minute"] = datetime_dim["full_datetime"].dt.minute
datetime_dim["weekday"] = datetime_dim["full_datetime"].dt.weekday
datetime_dim["weekday_name"] = datetime_dim["full_datetime"].dt.day_name()
datetime_dim["is_weekend"] = datetime_dim["weekday"] >= 5

datetime_dim
```

```python
datetime_dim.to_sql(
    "datetime_dim",
    conn,
    if_exists="append",
    index=False
)
```

```python
pu = df[["PULocationID"]].rename(columns={"PULocationID": "location_id"})
do = df[["DOLocationID"]].rename(columns={"DOLocationID": "location_id"})

locations = pd.concat([pu, do]).drop_duplicates()
```

```python
zones = pd.read_csv("taxi_zone_lookup.csv")
zones = zones.rename(columns={"LocationID": "location_id"})
location_dim = locations.merge(
    zones,
    on="location_id",
    how="left"
)
```

```python
location_dim = location_dim.dropna(subset=["Borough"])
location_dim.columns = location_dim.columns.str.lower()
```

```python
location_dim.to_sql(
    "location_dim",
    conn,
    if_exists="append",
    index=False
)
```
