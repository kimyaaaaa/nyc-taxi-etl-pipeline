```python
df = pd.read_parquet("data/raw/yellow_tripdata_2025-01.parquet")
df.columns = df.columns.str.lower()
df.info()
print(df.shape)
len(df)
```

```python
bad_rows = df[
    (df["trip_distance"] <= 0) |
    (df["tpep_dropoff_datetime"] <= df["tpep_pickup_datetime"])
]
df = df.drop(bad_rows.index)
df = df.reset_index(drop=True)
```

```python
def tod(h):
    if 5 <= h < 12: return "Morning"
    if 12 <= h < 17: return "Afternoon"
    if 17 <= h < 21: return "Evening"
    return "Night"

all_datetimes = pd.concat([
    df["tpep_pickup_datetime"],
    df["tpep_dropoff_datetime"]
]).drop_duplicates().sort_values()

datetime_dim = pd.DataFrame({
    "datetime_id": range(1, len(all_datetimes)+1),
    "full_datetime": all_datetimes
}).reset_index(drop=True)

datetime_dim["minute"] = datetime_dim["full_datetime"].dt.minute
datetime_dim["hour"] = datetime_dim["full_datetime"].dt.hour
datetime_dim["day"] = datetime_dim["full_datetime"].dt.day
datetime_dim["month"] = datetime_dim["full_datetime"].dt.month
datetime_dim["year"] = datetime_dim["full_datetime"].dt.year
datetime_dim["weekday"] = datetime_dim["full_datetime"].dt.weekday
datetime_dim["weekday_name"] = datetime_dim["full_datetime"].dt.day_name()
datetime_dim["is_weekend"] = datetime_dim["weekday"].isin([5,6])
datetime_dim["time_of_day"] = datetime_dim["hour"].apply(tod)
```

```python
datetime_dim.to_sql(
    "datetime_dim",
    engine,
    if_exists="append",
    index=False,
    chunksize=100_000,
    method="multi"
)
```

```python
def apply_location_scd(location_stage: pd.DataFrame, engine):
    """
    Applies SCD Type 2 logic to location_dim table.
    """

    now = datetime.utcnow()

    with engine.begin() as conn:

        for _, row in location_stage.iterrows():

            existing = pd.read_sql("""
                SELECT *
                FROM location_dim
                WHERE location_id = %s
                  AND is_current = true
            """, conn, params=[row["location_id"]])

            # CASE 1 — New location
            if existing.empty:

                conn.execute(text("""
                    INSERT INTO location_dim
                    (location_id, borough, zone, service_zone,
                     valid_from, valid_to, is_current)
                    VALUES (:id, :b, :z, :s, :from, NULL, true)
                """), {
                    "id": row["location_id"],
                    "b": row["borough"],
                    "z": row["zone"],
                    "s": row["service_zone"],
                    "from": now
                })

            else:
                old = existing.iloc[0]

                # CASE 2 — No change
                if (
                    old["borough"] == row["borough"]
                    and old["zone"] == row["zone"]
                    and old["service_zone"] == row["service_zone"]
                ):
                    # CASE 3 — Changed
                    conn.execute(text("""
                        UPDATE location_dim
                        SET is_current = false,
                            valid_to = :now
                        WHERE location_sk = :sk
                    """), {
                        "now": now,
                        "sk": old["location_sk"]
                    })

                conn.execute(text("""
                    INSERT INTO location_dim
                    (location_id, borough, zone, service_zone,
                     valid_from, valid_to, is_current)
                    VALUES (:id, :b, :z, :s, :from, NULL, true)
                """), {
                    "id": row["location_id"],
                    "b": row["borough"],
                    "z": row["zone"],
                    "s": row["service_zone"],
                    "from": now
                })
```

```python
pu = df[["pulocationid"]].rename(columns={"pulocationid": "location_id"})
do = df[["dolocationid"]].rename(columns={"dolocationid": "location_id"})

locations = pd.concat([pu, do]).drop_duplicates()

zones = pd.read_csv("taxi_zone_lookup.csv")
zones = zones.rename(columns={"LocationID": "location_id"})

location_dim = locations.merge(
    zones,
    on="location_id",
    how="left"
)

location_dim.columns = location_dim.columns.str.lower()

location_dim[["borough","zone","service_zone"]] = (
    location_dim[["borough","zone","service_zone"]]
    .fillna("Unknown")
)

location_dim["valid_from"] = pd.Timestamp.now('UTC')
location_dim["valid_to"] = None
location_dim["is_current"] = True

location_dim = location_dim[
    ["location_id","borough","zone","service_zone",
     "valid_from","valid_to","is_current"]
]

location_dim = location_dim.sort_values("location_id")
location_dim.reset_index(drop=True)
```

```python
location_dim.to_sql(
    "location_dim",
    engine,
    if_exists="append",
    chunksize=100_000,
    method="multi",
    index=False
)
```

```python
vendor_dim = (
    df[["vendorid"]]
    .drop_duplicates()
    .rename(columns={"vendorid": "vendor_id"})
)
vendor_names = {
    1: "Creative Mobile Technologies",
    2: "VeriFone"
}

vendor_dim["vendor_name"] = (
    vendor_dim["vendor_id"]
    .map(vendor_names)
    .fillna("Unknown")
)

vendor_dim.reset_index(drop=True)

vendor_dim.to_sql(
    "vendor_dim",
    engine,
    if_exists="append",
    index=False
)
```

```python
ratecode_dim = (
    df[["ratecodeid"]]
    .drop_duplicates()
    .dropna()
    .rename(columns={"ratecodeid": "ratecode_id"})
)

ratecode_map = {
    1: "Standard rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated fare",
    6: "Group ride"
}

ratecode_dim["rate_description"] = (
    ratecode_dim["ratecode_id"]
        .map(ratecode_map)
        .fillna("Unknown")
)

ratecode_dim["ratecode_id"] = ratecode_dim["ratecode_id"].astype(int)

ratecode_dim = ratecode_dim.sort_values("ratecode_id").reset_index(drop=True)

ratecode_dim.to_sql(
    "ratecode_dim",
    engine,
    if_exists="append",
    index=False
)
```

```python
trips_fact_df["passenger_count"] = trips_fact_df["passenger_count"].fillna(1)
trips_fact_df["congestion_surcharge"] = trips_fact_df["congestion_surcharge"].fillna(0)
trips_fact_df["airport_fee"] = trips_fact_df["airport_fee"].fillna(0)

# --- flags ---
trips_fact_df["store_and_fwd_flag"] = trips_fact_df["store_and_fwd_flag"].fillna("N")
trips_fact_df.isnull().sum()
```

```python
trips_fact_df = trips_fact_df[
    (trips_fact_df["trip_distance"] <= 1000) &
    (trips_fact_df["fare_amount"] <= 10000) &
    (trips_fact_df["total_amount"] <= 10000)
]
```

```python
trips_fact_df = trips_fact_df[
    (trips_fact_df["trip_duration_minutes"] >= 1) &
    (trips_fact_df["trip_duration_minutes"] <= 300)
]
money_cols = ["fare_amount","total_amount","tip_amount","tolls_amount","extra"]

for c in money_cols:
    trips_fact_df = trips_fact_df[trips_fact_df[c] >= 0]
```

```python
trips_fact_df["avg_speed_mph"] = (
    trips_fact_df["trip_distance"] /
    (trips_fact_df["trip_duration_minutes"] / 60)
)

trips_fact_df["avg_speed_mph"] = trips_fact_df["avg_speed_mph"].clip(0, 120)
```
